```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import os
import cv2
import tensorflow as tf
from bs4 import BeautifulSoup
from sklearn.model_selection import train_test_split
from PIL import Image
from keras.regularizers import l2
import glob
from keras.preprocessing.image import ImageDataGenerator
```

```python
# resize the image
IMAGE_SIZE = 224
```

# Read car images and resize them

```python
img_dir = 'rawData/images'  # Enter Directory of all images 
data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path)
files.sort() #We sort the images in alphabetical order to match them to the xml files containing the annotations of the bounding boxes
X=[]
for f1 in files:
    img = cv2.imread(f1)
    img = cv2.resize(img, (IMAGE_SIZE,IMAGE_SIZE))
    X.append(np.array(img))
    
```

```python
from lxml import etree
def resizeannotation(f):
    tree = etree.parse(f)
    for dim in tree.xpath("size"):
        width = int(dim.xpath("width")[0].text)
        height = int(dim.xpath("height")[0].text)
    for dim in tree.xpath("object/bndbox"):
        xmin = int(dim.xpath("xmin")[0].text)/(width/IMAGE_SIZE)
        ymin = int(dim.xpath("ymin")[0].text)/(height/IMAGE_SIZE)
        xmax = int(dim.xpath("xmax")[0].text)/(width/IMAGE_SIZE)
        ymax = int(dim.xpath("ymax")[0].text)/(height/IMAGE_SIZE)
    return [int(xmax), int(ymax), int(xmin), int(ymin)]
'rawData/annotations'
```

# Read licence plate annotations and resize them

```python
path = 'rawData/annotations'
text_files = ['rawData/annotations/'+f for f in sorted(os.listdir(path))]
y=[]
for i in text_files:
    y.append(resizeannotation(i))
```

```python
import keras

cnn = setupModel()

cnn.summary()
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

cnn.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

checkpoint = [
    EarlyStopping(monitor='val_loss' , mode='min', patience=50, verbose=1),
    ModelCheckpoint('model-{epoch:03d}.model',
                             monitor='val_accuracy',
                             verbose=0,
                             save_best_only=True,
                             mode='max')
]

train =cnn.fit(X_train, y_train, 
               validation_data=(X_val, y_val), 
                epochs=100, 
                batch_size=32,
                verbose=1,
                callbacks=[checkpoint])
```

```python
#Test
scores = cnn.evaluate(X_test, y_test, verbose=0)
print("Score : %.2f%%" % (scores[1]*100))
```

```python
plt.rcParams.update(plt.rcParamsDefault)

accuracy = train.history['accuracy']
val_accuracy = train.history['val_accuracy']
loss = train.history['loss']
val_loss = train.history['val_loss']
epochs = range(len(accuracy))
plt.plot(epochs, accuracy, 'b', label='Training set accuracy')
plt.plot(epochs, val_accuracy, 'r', label='Test set accuracy')
plt.plot(epochs, loss, 'g', label='Training set loss')
plt.plot(epochs, val_loss, 'y', label='Test set loss')
plt.title('Scores')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.yticks(np.linspace(0, 1, 11))

plt.legend()
plt.show()
```
